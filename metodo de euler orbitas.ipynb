```python
from numpy import array,cos,sin
from math import pi, sqrt, tan
from numpy import loadtxt
import numpy as np
import scipy as sc
import matplotlib.pyplot as plot
from pylab import plot, show,plt
```

```python
#us=1.495*10**11
m=1
gm=(4*pow(3.1416,2))
dt=0.02
n=200
r_o=1
v=2*(3.1416)
the=0
e_n = [] # M* AU^2/yr^2
x_pos = [] # UA
y_pos = [] # UA
r_pos = []
th_pos = []

def acce_i(r):
    return -gm / (r*r*r)

def ener(r):
    return -(gm*m) / (2*r)


```

```python
x_n = r_o
y_n = 0.
vx_n = 0.
vy_n = v

r_n = sqrt(x_n*x_n + y_n*y_n)
th = 0.

e_n.append( ener(r_n) )
x_pos.append( x_n )
y_pos.append( x_n )
r_pos.append( r_n )
th_pos.append( th )

for t in range(n):
    tmp_x = x_n
    tmp_y = y_n
    tmp_th = th
    
    vx_n = vx_n + tmp_x*acce_i(r_n)*dt
    vy_n = vy_n + tmp_y*acce_i(r_n)*dt
    
    x_n = tmp_x + vx_n*dt
    y_n = tmp_y + vy_n*dt
    
      
    r_n = sqrt(x_n*x_n + y_n*y_n)
    th = tmp_th + ( sqrt(vx_n*vx_n + vy_n*vy_n)/r_n ) * dt
        
    e_n.append( ener(r_n) )
    x_pos.append( x_n )
    y_pos.append( y_n )
    r_pos.append( r_n )
    th_pos.append( th )
```

```python
plt.plot(x_pos,y_pos,'o')
```

```python
plt.polar(th_pos,r_pos,'o')
plt.show()


plt.plot(e_n)
plt.show()
```
