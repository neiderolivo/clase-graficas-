```python
from numpy import linspace, sin, cos 
x= linspace(0, 10,1000)
y=sin(x)
z=cos(x)
plot(x,z)
plot(x,y)
show()
```

```python
from pylab import ylim, xlim, grid,xlabel,ylabel
x= linspace(0, 10,1000)
y=sin(x)


plot(x,y,'r--')# este es para imprimir normal
ylim(-1.1,1.1)
xlim(0.,10)
grid("on")
xlabel("eje x")
ylabel("eje y")
show()
```
